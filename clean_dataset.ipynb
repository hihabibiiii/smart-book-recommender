```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/books.csv")
# df.head()
```

```python
# Keep only needed columns
df = df[["original_title", "authors", "image_url"]]
```

```python
# Rename columns for recommender
df = df.rename(columns={
    "original_title": "name",
    "authors": "description"
})
```

```python
# Add category column
df["category"] = "book"
```

```python
# Drop missing titles
df = df.dropna(subset=["name"])
```

```python
# Reset index
df = df.reset_index(drop=True)
```

```python
df.to_csv("./data/products.csv", index=False)
print("Clean dataset saved as products.csv ✅")
```
